```python
import numpy, scipy, matplotlib.pyplot as plt, pandas, librosa,sklearn
def confusion_matrix(cm, classes,
                          baslik='Doğruluk Tablosu',
                          cmap=plt.cm.Oranges):
   
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(baslik)
    plt.colorbar()
    tick_marks = numpy.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('Asıl Değerler')
    plt.xlabel('Tahmin Edilen')

def plot_cnf(model,dataset_x,dataset_y,Turler):
    true_y=dataset_y
    true_x=dataset_x
    pred=model.predict(true_x)
    cnf_matrix=sklearn.metrics.confusion_matrix(true_y,pred)
    plt.figure()
    a=confusion_matrix(cnf_matrix,classes=Turler,baslik='Doğruluk Tablosu')
```

```python
%matplotlib inline
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
import numpy as np
import itertools
import itertools
import numpy, scipy, matplotlib.pyplot as plt, pandas, librosa,sklearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
results_knn=[]
```

# KNN algoritmasının uygulanması ve performans analizi

```python
#for i in range(1,11):
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train,y_train)
results_knn.append(knn.score(X_test,y_test))
```

```python
max_accuracy_knn=max(results_knn)
best_k=1+results_knn.index(max(results_knn))
print("Max Accuracy is {:.3f} on test dataset with {} neighbors.\n".format(max_accuracy_knn,best_k))
```

```python
knn=KNeighborsClassifier(n_neighbors=best_k)
knn.fit(X_train,y_train)
print("Knn Başarı Yüzdesi: {:.3f}".format(knn.score(X_test,y_test)))  
```

```python
plot_cnf(knn,X_test,y_test,Turler)
# print(y_test)
```

# Random Forest algoritmasının uygulanması ve performans analizi

```python
results_forest=[]
# for i in range(2,20):
forest=RandomForestClassifier(random_state=42,n_estimators=20)
forest.fit(X_train,y_train)
results_forest.append(forest.score(X_test,y_test))
    
```

```python
max_accuracy_forest=max(results_forest)
best_n_est=2+results_forest.index(max(results_forest))

print("Max Accuracy is {:.3f} on test dataset with {} estimators.\n".format(max_accuracy_forest,best_n_est))
```

```python
forest=RandomForestClassifier(random_state=42,n_estimators=best_n_est)
forest.fit(X_train,y_train)
print("RandomForest Başarı Yüzdesi: {:.3f}".format(forest.score(X_test,y_test)))

plot_cnf(forest,X_test,y_test,Turler)
```
